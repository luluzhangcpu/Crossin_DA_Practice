```python
# task5--数据的选取（索引方式)
# 任务提示：对 DataFrame 进行选择，主要有以下几种索引方式：
# ① 使用 [ ]；
# ② 使用 df.loc[ ]；
# ③ 使用 df.iloc[ ]，此种方式为标签位置索引，感兴趣的可自行了解；
# task5-1--使用[ ]位置索引方式，选取出前 3 行数据
print(df1.iloc[:3,:])
print('-'*80)
print(df1[:3])
```

```python
# task5-2--使用[ ]标签索引方式，选取出 “date” 一列，并用 type 查看返回的数据类型
a = df1.loc[:,'date']
print(a)
print('-' * 30)
print(type(a))
```

```python
# task5-3--使用[ ]标签索引方式，取出 “date”、“open” 列和 “colse” 列，
# 并用 type 查看返回的数据类型
b = df1.loc[:,['date','open','close']]
print(b)
print('-' * 30)
print(type(b))
```

```python
# task5-4--使用[ ]标签索引方式，选取收盘价大于 11.08 的所有数据
c = df1[df1['close']>11.08]
print(c)
```

```python
# task5-5--将 date 设置为索引，并重新命名数据库为 df1
df1.set_index('date',inplace=True)
print(df1)
```

```python
# 使用 df.loc[] 标签定位方式，
# 选取 df1 数据库中 2017-12-11 至 2017-12-20 的所有数据
d = df1.loc['2017-12-11':'2017-12-20']
print(d)
```

```python
# task5-6--使用df.loc[]标签定位方式，选取 df1 数据库中 open 至 close 间的所有列
e = df1.loc[:,'open':'close']
print(e)
```

```python
# task 5-7--使用 df.loc[] 标签定位方式，
# 选取 df1 数据库中时间从 2018-07-01 至 2018-07-08，open 至 close 间的所有列
f = df1.loc['2018-07-01':'2018-07-08','open':'close']
print(f)
```
